```python
# Exercise 1

# Let A = 4 and w1, w2 be two normal distributions . 
# For each of the cases below:
# - plot the PDF
# - Check in which of the two distributions we have a higher probability of a sample being equal to A (drawing A) (Hint: PDF again)
# The cases are:
# a) w1: [mean = 3, sigma = 1.5], w2: [mean = 7, sigma = 1]. 
# b) w1: [mean = 3, sigma = 1], w2: [mean = 5, sigma = 2]
# c) w1: [mean = 4, sigma = 1], w2: [mean = 4, sigma = 0.5]

from scipy.stats import norm #https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html
from matplotlib import pyplot as plt
import numpy as np

A = 4
x = np.arange(-3, 12, 0.001)  # Plotting interval

w1a = norm(3, 1.5)
w2a = norm(7, 1)
w1b = norm(3, 1)
w2b = norm(5, 2)
w1c = norm(4, 1)
w2c = norm(4, 0.5)

plt.plot(x, w1a.pdf(x), label="w1a")
plt.plot(x, w2a.pdf(x), label="w2a")
plt.title("Case a")
plt.legend()
plt.show()
plt.plot(x, w1b.pdf(x), label="w1b")
plt.plot(x, w2b.pdf(x), label="w2b")
plt.legend()
plt.title("Case b")
plt.show()
plt.plot(x, w1c.pdf(x), label="w1c")
plt.plot(x, w2c.pdf(x), label="w2c")
plt.legend()
plt.title("Case c")
plt.show()

print(f"Case a: pdf w1: {w1a.pdf(A)}, pdf w2: {w2a.pdf(A)}")
print(f"Case b: pdf w1: {w1b.pdf(A)}, pdf w2: {w2b.pdf(A)}")
print(f"Case c: pdf w1: {w1c.pdf(A)}, pdf w2: {w2c.pdf(A)}")
```

```python
# Exercise 2 

# Let X, y represent a labelled dataset, where X represents the input features and y the target (aka label, ground truth)
# Let two feature sets A and B be defined as the points belonging to class 0 or 1 respectively.
# Find the corresponding centroids centroids of each set (x,y), then calculate the Euclidean distance between them and each proposed z.
# Produce the scatter plot of the points of A and B, the centroids (using a different marker from the one used to plot the points) 
# and the point z for the various cases. To which class would you assign z in each case? Why?
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html
# a) z = [6.5, 3]
# b) z = [4.5, 3.5]
# c) z = [5.5, 3.25]

import numpy as np
from sklearn import datasets
from scipy.spatial import distance
from matplotlib import pyplot as plt

iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features (It's easy to do scatterplots when working in 2D).
y = iris.target

A = X[y == 0]
B = X[y == 1]

ca = [np.mean(A[:,0]), np.mean(A[:,1])]
cb = [np.mean(B[:,0]), np.mean(B[:,1])]
print("A centroid:",ca)
print("B centroid:",cb)

z1 = [6.5, 3]
z2 = [4.5, 3.5]
z3 = [5.5, 3.25]
print("#"*50)
plt.scatter(A[:,0], A[:,1], label="A",c="g")
plt.scatter(B[:,0], B[:,1], label="B",c="b")
plt.scatter(ca[0], ca[1], marker="X",s=100,c="r", label="ca")
plt.scatter(cb[0], cb[1], marker="+",s=100,c="r", label="cb")
plt.scatter(z1[0], z1[1], marker="*",s=100,c="r", label="z1")
plt.scatter(z2[0], z2[1], marker="o",s=100,c="r", label="z2")
plt.scatter(z3[0], z3[1], marker="v",s=100,c="r", label="z3")
plt.legend()
plt.show()

print("#"*50)
print("Euclidean distance between z1 and A centroid:",distance.euclidean(ca,z1))
print("Euclidean distance between z1 and B centroid:",distance.euclidean(cb,z1))
print("Euclidean distance between z2 and A centroid:",distance.euclidean(ca,z2))
print("Euclidean distance between z2 and B centroid:",distance.euclidean(cb,z2))
print("Euclidean distance between z3 and A centroid:",distance.euclidean(ca,z3))
print("Euclidean distance between z3 and B centroid:",distance.euclidean(cb,z3))

# TODO
```
